```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(20)

cities = ["Tehran", "Mashhad", "Isfahan", "Tabriz", "Shiraz", "Ahvaz"]
days = np.arange(1, 31)

all_data = []

for city in cities:
    temp = np.random.randint(15, 41, size=30)
    hum = np.random.randint(20, 81, size=30)
    rain = np.random.randint(0, 51, size=30)

    df_city = pd.DataFrame({
        "City": city,          
        "Day": days,
        "Temperature": temp,
        "Humidity": hum,
        "Rainfall": rain
    })

    all_data.append(df_city) 


df = pd.concat(all_data, ignore_index=True)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
print("----------------show data frame---------")
display(df)
```

```python

#Question1-section3-Q1
print("""
      ----------Question1-section3-q1---------""")
avg_temp_per_city=df.groupby("City")["Temperature"].mean()
print("Average of Temperature for each city:",avg_temp_per_city)

avg_hum_per_city=df.groupby("City")["Humidity"].mean()
print("Average of Humidity for each city:",avg_hum_per_city)

avg_rain_per_city=df.groupby("City")["Rainfall"].mean()
print("Average of Rainfall for each city:",avg_rain_per_city)



```

```python
#Question1-section3-Q2
print("""
      ----------Question1-section3-q2---------""")
max_temp=avg_temp_per_city.max()
hottest_city=avg_temp_per_city[avg_temp_per_city==max_temp].index[0]
print("The hottest city:",hottest_city,"Max temp:",max_temp)

min_temp=avg_temp_per_city.min()
coldest_city=avg_temp_per_city[avg_temp_per_city==min_temp].index[0]
print("The coldest city:",coldest_city,"Min temp:",min_temp)

print("""
      ----------Question1-section3-q3-----------""")

rain_above_10=df[df["Rainfall"]>10]
print("Count of rainy days:",len(rain_above_10))
rain_above_10_per_city=df[df["Rainfall"]>10].groupby("City")["Rainfall"].count()
print("Groupby rainfall>10 :",rain_above_10_per_city)
```

```python
print("""
      ----------Question1-section4-----------""")

Isfahan_temp=df[df["City"]=="Isfahan"]["Temperature"]
Isfahan_hum=df[df["City"]=="Isfahan"]["Humidity"]

plt.scatter(Isfahan_temp,Isfahan_hum)
plt.title("Temperature and Humidity Relationship-Isfahan(30 Days)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity(%)")
plt.show()
print("""
      ----------Question1-section5-----------""")
print("No relationship between Temperature and Humidity")
```

```python
df2=pd.read_csv("diabetes.csv",nrows=None)
display(df2)
```

```python
#Question2-section1-Q1
print("""
      ----------Question2-section1-Q1-----------""")
colms_name=df2.columns
for item in colms_name:
    colm_mean=round(df2[item].mean(),3)
    df2[item].fillna(colm_mean,inplace=True)

display(df2)    
```
